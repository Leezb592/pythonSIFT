```python




def mkG(img,n):
   # 制造高斯空间 
    Gspace=[]
    g=img.copy()
    Gspace.append(g)
    for i in range(n):
        g = cv.pyrDown(g)
        Gspace.append(g)
    return Gspace

```

```python


def FindKeypoint(space,idx):
    #寻找关键点
    Keypoint = []
    shape=[len(space),len(space[0]),len(space[0][0])]
    for i in range(1,shape[0]-1):
        for j in range(1,shape[1]-1):
            for k in range(1,shape[2]-1):
                if ifkeypoint(i,j,k,space,shape) and ifedge(i,j,k,space):
                    Keypoint.append([i,j*idx,k*idx])
    return Keypoint

```

```python


def ifkeypoint(i,j,k,space,shape):

    dx=[0,1,-1]
    dy=[0,1,-1]
    dz=[0,1,-1]
    flag=0
    dnow=space[i][j][k]
    for ix in range(3):
        for iy in range(3):
            for iz in range(3):
                if flag==0 :
                    flag=1
                    continue
                if space[i+dx[ix]][j+dy[iy]][k+dz[iz]] >= dnow :
                    return False
    return True

```

```python
def ifedge(i,j,k,space):
    dx=space[i][j+1][k]+space[i][j+1][k-1]+space[i][j+1][k+1]
    -(space[i][j-1][k]+space[i][j-1][k-1]+space[i][j-1][k+1])
    dx=abs(dx)
    dy=space[i][j+1][k+1]+space[i][j][k+1]+space[i][j-1][k+1]
    -(space[i][j+1][k-1]+space[i][j][k-1]+space[i][j-1][k-1])
    dy=abs(dy)
    if dy==0 or dx==0: return false
    t=1.0*dx/dy
    if t<10 and t>0.1: 
        return True
    else :
        return False
    
    
```

```python

def extract_features(image_paths):
    sift_det = cv2.xfeatures2d.SIFT_create()
    imgDes_list = []  # 所有图像的所有sift descriptor的列表
    for i, image_path in enumerate(image_paths):
        img = cv2.imread(image_path)
        gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        kp, img_des = sift_det.detectAndCompute(gray, None) 
# 以第一幅图像为例，kp是这幅图像的所有keypoints。img_des是这幅图像的keypoint descriptor的列表，
#它的shape为(3765, 128)。即第一幅图像中提取到了3765个关键点，每个关键点对应一个描述符，而一个描述#符是一个128维的向量。
        imgDes_list.append(img_des)
    return imgDes_list
```
